```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
import time
```

```python
#crawl each unit
def unit(article, driver):
    link = article.find_element_by_class_name("timeline__item__title")
    link.click()
    time.sleep(3)
    driver.switch_to_window(driver.window_handles[-1])
    title = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[2]/div[2]/article/h1")))
    body = driver.find_element_by_class_name("article__bd__detail")
    newsList = body.find_elements_by_tag_name("p")[2:]
    newsPositive = pd.DataFrame({'Title': [], 'Positive': []})
    newsNegative = pd.DataFrame({'Negative': []})
    newsTitle = pd.DataFrame({'Title': []})
    newsCollect = []
    for i, news in enumerate(newsList):
        string = ""
        for piece in news.text:
            string += piece.encode('utf-8').strip()
        string = string.strip()
        if string == "————负面新闻————":
            #print 'into negative'
            newsPositive.Positive = newsCollect
            newsCollect = []
            continue
        #print string.encode("utf-8")
        newsCollect.append(string)
        #print '\n'
    newsNegative.Negative = newsCollect
    unitResult = pd.concat([newsPositive, newsNegative], axis = 1)
    unitResult.Title = title.text.encode('utf-8')
    driver.close()
    driver.switch_to_window(driver.window_handles[-1])
    #print unitResult
    return unitResult
```

```python
#main flow
outfile = open('snowball.csv', 'a')
driver = webdriver.Firefox()
driver.set_window_size(1000, 1000)
url = "https://xueqiu.com/u/1689987310"
keyword = u"正负面新闻"
wait = WebDriverWait(driver, 10)
driver.get(url)
time.sleep(5)
search_box = driver.find_element_by_xpath("/html/body/div[2]/div[3]/div[1]/div[1]/div[3]/input")
search_box.send_keys(keyword + Keys.ENTER)
time.sleep(5)
while True:
    pageData = pd.DataFrame({'Title': [], 'Positive': [], 'Negative': []})
    element = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[2]/div[3]/div[1]/div[1]/div[1]")))
    nextPage = driver.find_elements_by_xpath("/html/body/div[2]/div[3]/div[1]/div[3]/a")[-1]
    articleList = driver.find_elements_by_tag_name("article")
    for article in articleList:
        try:
            result = unit(article, driver)
            pageData = pd.concat([pageData, result])
            time.sleep(3)
        except Exception as e:
            print e
    pageData.to_csv(outfile)
    nextPage.click()
```
